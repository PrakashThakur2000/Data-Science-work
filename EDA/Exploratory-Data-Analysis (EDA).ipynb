```python
#outlier detection using caping Method 
#first we define the function to find the upper and lower extreem

def outlier_detection(data,column):
    q1=data[column].quantile(0.25)
    q3=data[column].quantile(0.75)
    IQR=q3-q1
    
    lower=q1-(1.5*IQR)
    upper=q3+(1.5*IQR)
    
    return lower,upper
```

```python
#check the lower and upper extreem values
outlier_detection(df,"Wind")
```

```python
#Which values are greter than 17.65 means upper extreem

df[df['Wind']>17.65]
```
